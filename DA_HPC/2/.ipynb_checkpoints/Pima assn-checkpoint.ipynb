```python
import numpy as np
```

```python
df = data.copy()
```

dq = df.iloc[:,1:8]

```python
for i in df.columns[1:8]:
    df[i].replace(0,np.nan,inplace = True)
    m = df[i].mean()
    df[i].replace(np.nan,m,inplace = True)
```

```python
df.describe(include = "all")
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = np.array(df.drop('outcome',axis = 1))
```

```python
y = np.array(df['outcome'])
```

```python
s = StandardScaler()
```

```python
x_scale = s.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scale,y,random_state = 0,test_size = 0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB()
```

```python
gnb.fit(x_train,y_train)
```

```python
y_pred = gnb.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import confusion_matrix
```

```python
conf_mat = confusion_matrix(y_test, y_pred)
```

```python
conf_mat
```

```python
acc_scr = accuracy_score(y_test, y_pred)
```

```python
acc_scr
```

```python
from sklearn.metrics import precision_recall_fscore_support
```

```python
prf = precision_recall_fscore_support(y_test, y_pred)
```
